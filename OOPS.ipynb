```python
class Car:
    def __init__(self, window, door, enginetype):
        self.windows = window
        self.doors = door
        self.enginetype = enginetype
    # method
    def self_driving(self):
        return "This is a {} engine car".format(self.enginetype)
```

```python
car1 = Car(4, 5, "Petrol")
```

```python
print(car1.enginetype)
```

```python
car1.self_driving()
```

```python
car2 = Car(3, 4, "EV")
```

```python
print(car2.enginetype)
```

```python
class Student:

    college = "ABES"
    
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name: ", self.name)
        print("Age: ", self.age)
        print("College: ", Student.college)

    def greet(self):
        return "Hello {}!".format(self.name)
```

```python
# object
s1 = Student("Disha", 21)
```

```python
print(s1.age)
```

```python
print(s1.name)
```

```python
s1.display()
```
