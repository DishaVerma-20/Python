# Scatter Plot


```python
x = np.arange(0, 10)
y = np.arange(11, 21)
```

```python
## Scatter plot
plt.scatter(x, y, c = 'g' ,alpha = 0.5)
plt.scatter(x, y,c = 'r',alpha = 0.5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Graph in 2D")
plt.savefig('Test1.png')
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title("sine wave form")
plt.plot(x, y)
plt.show()
```

```python
# creating subplots
x = np.arange(0, 5*np.pi, 0.4)
y_sin = np.sin(x)
y_cos = np.cos(x)

# subplot grid of height = 2 and width = 1
plt.subplot(2, 1, 1)
plt.plot(x, y_sin, 'ro-', linewidth = 1)
plt.title("sine")
plt.subplot(2, 1, 2)
plt.plot(x, y_cos, 'g*-', linewidth = 2)
plt.title("cosine")
```

# BarPlot

```python
x = [2, 8, 10]
y = [11, 16, 9]

x2 = [3, 9, 11]
y2 = [6, 15, 7]
plt.bar(x, y, align = "center")
plt.bar(x2, y2, color = 'g')
plt.title("Bar Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

# Histogram

```python
import matplotlib.pyplot as plt

marks = [45,50,52,55,60,62,65,70,75,80]

plt.hist(
    marks,
    bins=5,
    color = 'c',
    edgecolor = 'b'
    # color='skyblue',
    # edgecolor='black'
)

plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

# Box Plot using Matplotlib

```python
data = [np.random.normal(0, std, 100) for std in range (1, 4)]
```

```python
# plt.boxplot(data, vert= True, patch_artist = True)
plt.boxplot(data, orientation = 'vertical', patch_artist = True)
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']

# Highlight the first slice (Python)
explode = (0.4, 0, 0, 0)

# Create pie chart
plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=False
)

# Make it a perfect circle
plt.axis('equal')

# Title
plt.title('Programming Language Popularity')

plt.show()
```
