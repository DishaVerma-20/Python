# custom exception

```python
class Error(Exception): # inheriting exception class in error class
    pass

class dobException(Error):
    pass
```

```python
year = int(input("Enter the year: "))
age = 2026-year
try:
    if age <= 30 and age > 20:
        print("The age is valid, you can apply for exam.")
    else:
        raise dobException
except dobException:
    print("Age is not valid")
```
