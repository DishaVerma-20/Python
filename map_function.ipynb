# map function

```python
def evenodd(num):
    if num % 2 == 0:
        return "The number {} is even".format(num)
    else:
        return "The number {} is odd".format(num)
```

```python
evenodd(8)
```

```python
lst = [8, 5, 6, 3, 5, 6, 9, 67, 56]
```

```python
map(evenodd, lst)
```
