```python
def greet(name):
    return "Hello {}, welcome to the community!".format(name)
```

```python
greet("Disha Verma")
```

```python
def welcome_email(name, age):
    return "Welcome {} aged {}!".format(name, age)
```

```python
welcome_email("Disha", 21)
```

```python
# we can use placeholders, temporarily variables
def welcome_email(name, age):
    return "Welcome {name} aged {age}!".format(age = age, name = name)
```

```python
welcome_email("Disha", 21)
```

```python
# different placeholder name
def welcome_email(name, age):
    return "Welcome {name1} aged {age1}!".format(age1 = age, name1 = name)
```

```python
welcome_email("Disha", 21)
```
