```python
def even (num):
    if num%2==0:
        # print("The number {} is even".format(num))
        return True
```

```python
even (8)
```

```python
lst = [1,2,3,4,5,6,7,8,9]
```

```python
filter(even, lst)
```

```python
list(filter(even, lst))
```

```python
map(even, lst)
```

```python
list(map(even, lst))
```
