```python
# take number as input from the user one by one. Skip negative numbers and keep adding positive numbers. stop when the user enters 0 and print the total.
```

```python
sum = 0
while True:
    num = int(input("Enter the number:- "))
    if num > 0:
        sum = sum + num
    elif (num == 0):
        print(sum)
        break
    else: 
        continue
```
