# Lambda Function
Anonymous Function.
A function with no name

```python
def addition(a, b):
    return a+b
# single line function so convert it into lambda function
```

```python
addition = lambda a,b:a+b
```

```python
addition(2,3)
```

```python
def even(num):
    if num%2==0:
        return True
```

```python
even(24)
```

```python
even = lambda a:a%2==0
```

```python
even(45)
```

```python
def add(x,y, z):
    return x+y+z
```

```python
add(4,5,7)
```
