# Exception Handling

```python
a = b
```

```python
try:
    # code block where exception can occur
    a = b
except:
    print("Some problem may have occured")
```

```python
try:
    a = b
except NameError as ex:
    print("The user has not defined the variable")
except Exception as ex:
    print(ex)
```

```python
try:
    a = 1
    b = 's'
    c = a + b
except NameError as ex1:
    print("The user has not defined the variable")
except TypeError as ex2:
    print("Try to make the datatype similar")
except ZeroDivisionError:
    print("can't divide any number by 0")
except Exception as ex:
    print(ex)
```

```python
try:
    a = int(input("Enter the number 1: "))
    b = int(input("Enter the number 2: "))
    c = a/b
    d = a*b
    e = a+b
except NameError as ex1:
    print("The user has not defined the variable")
except ZeroDivisionError as ex2:
    print("Division ny zero is not possible")
except TypeError as ex3:
    print("Try to make the datatype similar")
except Exception as ex:
    print(ex)
else:
    print(c)
    print(d)
    print(e)
# if exception is getting handled, it willn't touch the else block
```

```python
try:
    a = int(input("Enter the number 1: "))
    b = int(input("Enter the number 2: "))
    c = a/b
except NameError as ex1:
    print("The user has not defined the variable")
except ZeroDivisionError as ex2:
    print("Division ny zero is not possible")
except TypeError as ex3:
    print("Try to make the datatype similar")
except Exception as ex:
    print(ex)
else:
    print(c)
finally:
    print("The execution is done!")
```

```python
try:
    a = int(input("Enter the number 1: "))
    b = int(input("Enter the number 2: "))
    c = a/b
except NameError as ex1:
    print("The user has not defined the variable")
except ZeroDivisionError as ex2:
    print("Division ny zero is not possible")
except TypeError as ex3:
    print("Try to make the datatype similar")
except Exception as ex:
    print(ex)
else:
    print(c)
finally:
    print("The execution is done!")
# finally is used to close the connection
```
