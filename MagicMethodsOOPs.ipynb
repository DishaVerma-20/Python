# Python OOPS - Magic Method in classes

```python
class Car():
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
    def drive(Self):
        print("The person drives the car.")
```

```python
car1 = Car(4, 5, "EV")
```

```python
dir(car1)
```

```python
class Car():
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
    def __str__(self):
        print("The object has been initialized")
    def drive(Self):
        print("The person drives the car.")
```

```python
car1.__str__()
```

```python
print(car1.__str__())
```

```python
print(car1)
```

```python
class Car():
    def __new__(Self, windows, doors, enginetype):
        print("The object has started getting initialized")
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
    def __sizeof__(self):
        return "This displays the size of the object"
    def __str__(self):
        return "The object has been initialized"
    def drive(Self):
        print("The person drives the car.")
```

```python
car1 = Car(4, 5, "EV")
```

```python
print(car1)
```

```python
car1.__sizeof__()
```

```python
# Magic methods allow us to customize how our objects behave when Python performs built-in operations on them.
# Magic methods (dunder methods) are special methods that Python automatically invokes to define the behavior of objects for built-in operations 
# like object creation, printing, addition, length calculation,etc.
# Python ke objects ke peeche bahut saare hidden methods (magic methods) chal rahe hote hain.
```
